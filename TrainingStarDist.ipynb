```python
from __future__ import print_function, unicode_literals, absolute_import, division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from glob import glob
from tqdm import tqdm
from tifffile import imread
from csbdeep.utils import Path, normalize
from csbdeep.utils import axes_dict, plot_some, plot_history
from stardist import fill_label_holes, random_label_cmap
from stardist import Config, StarDist, StarDistData

np.random.seed(42)
lbl_cmap = random_label_cmap()
```

# Specify the directory of training images. The number of ground truth images must be the same name, size as the mask images. Given the size of the input images choose an appropriate number of rays needed for the stardist mode. If the cells to be segmented are roundish N_rays = 32 should be enough, for describing a more complex convex shape cells choose a higher value of the number of rays, such as 128 or 512.

```python
X = sorted(glob('/data/u934/service_imagerie/v_kapoor/from_Ozge_for_curvature/multiple_cells/images/*.tif'))
Y = sorted(glob('/data/u934/service_imagerie/v_kapoor/from_Ozge_for_curvature/multiple_cells/masks/*.tif'))
assert all(Path(x).name==Path(y).name for x,y in zip(X,Y))
```

```python
X = list(map(imread,X))
Y = list(map(imread,Y))


N_rays = 32
patchSize = (1024,1024)
ModelName = 'EmbryoCell32Rays'
ModelDir = '../models/Curiemodels'
```

Specify the hyperparameters for the custom U-net model

```python
trainEpochs = 50
Steps_per_epoch = 200 
FilterNumber = 32
NetworkDepth = 3 
batch_size = 1
```

```python
X = [normalize(x,1,99.8) for x in tqdm(X)]
Y = [fill_label_holes(y.astype(int)) for y in tqdm(Y)]


```

```python
rng = np.random.RandomState(42)
ind = rng.permutation(len(X))
n_val = int(round(0.15 * len(X)))
ind_train, ind_val = ind[:-n_val], ind[-n_val:]
X_val, Y_val = [X[i] for i in ind_val]  , [Y[i] for i in ind_val]
X_trn, Y_trn = [X[i] for i in ind_train], [Y[i] for i in ind_train] 



print('number of images: %3d' % len(X))
print('- training:       %3d' % len(X_trn))
print('- validation:     %3d' % len(X_val))
print(X[0].shape[0])
```

```python
np.random.seed(42)
data = StarDistData(X,Y,batch_size=1,n_rays=N_rays,patch_size=patchSize,shape_completion=False)
```

```python
(img,dist_mask), (prob,dist) = data[0]

fig, ax = plt.subplots(2,2, figsize=(12,12))
for a,d,cm,s in zip(ax.flat, [img,prob,dist_mask,dist], ['gray','magma','bone','viridis'],
                    ['Input image','Object probability','Distance mask','Distance (0°)']):
    a.imshow(d[0,...,0],cmap=cm)
    a.set_title(s)
plt.tight_layout()
None;
```

```python
conf = Config(train_batch_size=batch_size, train_shape_completion=False, train_patch_size = patchSize, train_epochs = trainEpochs, 
              train_steps_per_epoch = Steps_per_epoch, n_rays = N_rays, unet_n_filter_base = FilterNumber, unet_n_depth = NetworkDepth )
print(conf)
vars(conf)
```

```python
model = StarDist(conf, name=ModelName, basedir=ModelDir)
```

Based on the parameters and hyperparameters of the model start the training, for a proepr training the loss and validation loss must show the same trend. If the validation loss stops to decrease while the loss value is still decreasing, it indicates overfitting on the training data and hence the training data must be modified to ensure a proper training or the hyperparameters were not properly chosen for the given training data

```python
history = model.train(X_trn,Y_trn,validation_data=(X_val,Y_val))
```

```python
print(sorted(list(history.history.keys())))
plt.figure(figsize=(16,5))
plot_history(history,['loss','val_loss'],['prob_loss','val_prob_loss','dist_loss','val_dist_loss']);
```
